```python
import pandas as pd
pop = pd.read_csv("popmonth.csv", encoding='cp949', index_col=0, thousands=',') # index_col: 행의 제목, thousands=',': 천 단위 구분 기호를 제거하고 숫자로 인식
pop
```

```python
pop.columns = ['총인구수', '세대수', '세대당 인구 수', '남자 인구 수', '여자 인구 수', '남녀비율']
pop.head
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(5, 3))
sns.scatterplot(pop['총인구수'])
plt.xticks(rotation=90) # x의 문자 각도를 90도
plt.show()
```

```python
pop.sort_values('총인구수')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(5,3))

pop1 = pop.sort_values('총인구수')
sns.barplot(x=pop1.index, y='총인구수', data=pop1)
plt.xticks(rotation=90)
plt.show()
```
