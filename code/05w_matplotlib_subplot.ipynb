```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(-100, 100, 1000)
y = np.random.randint(-100, 100, 1000)

mask1 = abs(x) > 50
mask2 = abs(y) > 50
x = x[mask1 + mask2]  # mask1과 mask2 중 하나라도 만족하는 값 저장
y = y[mask1 + mask2]

plt.scatter(x, y, c=x, cmap='jet', alpha = 0.4) # cmap='jet' 컬러맵 값이 커질 수록 색 변경
plt.colorbar()
plt.show()
```

## plt.subplot()

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(9, 4))

plt.subplot(2, 3, 1)
plt.plot([1, 2, 3])
plt.subplot(2, 3, 5)
plt.subplot(2, 3, 6)
plt.plot([5, 4, 3])

plt.show()
```

```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(2, 4) # 2행 4열의 서브플롯
fig.set_size_inches((9, 4)) # 캔버스의 크기 지정

axs[0, 1].plot([5, 1, 3])
axs[1, 2].plot([3, 3, 1])
plt.plot([10, 5, 20])
plt.show()
```

```python
# plt.subplots()에서 직접 캔버스의 크기 지정
import matplotlib.pyplot as plt

# plt.subplots()로 figure와 크기를 지정
fig, axes = plt.subplots(2, 1, figsize=(8, 6))

# 각 서브플롯에 데이터 추가
axes[0].plot([1, 2, 3], [1, 4, 9])
axes[1].plot([1, 2, 3], [1, 2, 3])

plt.show()
```

```python
import matplotlib.pyplot as plt

# 전체 제목 - 어느 위치에서 호출해도 무관 (Figure 레벨)
plt.suptitle('전체 제목', fontsize=14, y=1.02)

# subplot 1 - 생성 직후 바로 title 지정
plt.subplot(2, 3, 1)
plt.plot([1, 4, 9, 16])
plt.title('제곱')

# subplot 5
plt.subplot(2, 3, 6)
plt.plot([1, 2, 3, 4])
plt.title('선형')

# subplot 2
plt.subplot(2, 3, 2)
plt.plot([6, 5, 2, 1])
plt.title('감소')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(2, 3, figsize=(7, 4))
fig.set_size_inches(6, 3)

ax[0, 1].plot([1, 4, 9, 16])
ax[1, 1].plot([1, 2, 3, 4])
ax[1, 2].plot([6, 5, 2, 1])

# 전체 제목 Figure level
fig.suptitle('전체 제목', fontsize=14, fontweight='bold')

# 각 subplot 제목 - Axes level
ax[0, 0].set_title('(0, 0) 빈 플롯')
ax[0, 1].set_title('(0, 1) 제곱수')
ax[0, 2].set_title('(0, 2) 빈 플롯')
ax[1, 0].set_title('(1, 0) 선형')
ax[1, 1].set_title('(1, 1) 빈 플롯')
ax[1, 2].set_title('(1, 2) 감소')

plt.tight_layout()
plt.show()
```
