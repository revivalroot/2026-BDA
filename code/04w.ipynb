### 1차원 배열 다양한 방향으로 쌓기

```python
import numpy as np

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
```

```python
# 수평 방향으로 쌓기
np.hstack((a, b))
```

```python
np.r_[a, b]
```

```python
np.concatenate((a, b))
```

```python
np.c_[a, b]
```

```python
import numpy as np

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
```

```python
# 수직 방향으로 쌓기
np.vstack((a, b))
```

```python
np.row_stack((a, b))
```

```python
np.r_[a[np.newaxis, :], b[np.newaxis, :]]
```

```python
np.concatenate((a[np.newaxis, :], b[np.newaxis, :]))
```

```python
import numpy as np

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
```

```python
np.column_stack((a, b))
```

```python
np.hstack((a[:, np.newaxis], b[:, np.newaxis]))
```

```python
np.c_[a, b]
```

```python
np.concatenate(((a[:, np.newaxis], b[:, np.newaxis])), axis=1)
```

## numpy의 난수와 다양한 함수

```python
r = rg.random() # [0, 1) 실수 난수
r
```

```python
rg.random((4, 3))
```

```python
import numpy as np

rg = np.random.default_rng(12345)
r = rg.integers(low=0, high=10, size=3)
r
```

```python
rg.integers(1, 7, (3, 4)) # 3행 4열에서 랜덤한 수 1~6까지가 들어감
```
